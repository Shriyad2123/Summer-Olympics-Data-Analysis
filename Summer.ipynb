## Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
```

## 1. In how many cities Summer Olympics is held so far?

```python
city = len(df['City'].unique())
print(city)
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold = df[df['Medal']=='Gold']
gold['Sport'].value_counts().head().plot.bar()
gold['Sport'].value_counts().head()
```

## 6. In which year India won first Gold Medal in Summer Olympics?

```python
for i in df['Year'].unique():
    print(i,len(df.loc[(df['Year']==i) & (df['Country']=='IND') & (df['Medal']=='Gold')]))
```

## 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for j in df['Event'].unique():
    data.append([j, len(df[df['Event']==j])])
    
pd.DataFrame(data,columns=['Event','numberOfPlayer']).sort_values(by='numberOfPlayer',ascending=False).head().plot.bar(x='Event',y='numberOfPlayer')
```

## 8. Which sport is having most female Gold Medalists? (Top 5)

```python
w = df[df['Gender'] == 'Women']
wg = w[w['Medal'] == 'Gold']
wg.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar()
```
